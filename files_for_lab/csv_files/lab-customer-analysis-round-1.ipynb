```python
import pandas as pd

df1 = pd.read_csv('file1.csv')
display(df1.head())

df2 = pd.read_csv('file2.csv')
display(df2.head())

df3 = pd.read_csv('file3.csv')
display(df3.head())
```

```python
print("here is the dataframe 1 shape: ", df1.shape)
print("here is the dataframe 1 shape: ", df2.shape)
print("here is the dataframe 1 shape: ", df3.shape)
```

```python
df1.rename(columns={'ST':'STATE'}, inplace=True)
df2.rename(columns={'ST':'STATE'}, inplace=True)

df1.columns = df1.columns.str.upper().str.replace(' ', '_')
df2.columns = df2.columns.str.upper().str.replace(' ', '_')
df3.columns = df3.columns.str.upper().str.replace(' ', '_')


display(df1.head())
display(df2.head())
display(df3.head())
```

```python
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
df3 = pd.DataFrame(df3)

desired_order = ['CUSTOMER', 'STATE', 'CUSTOMER_LIFETIME_VALUE', 'GENDER', 'EDUCATION', 'INCOME', 'MONTHLY_PREMIUM_AUTO', 'NUMBER_OF_OPEN_COMPLAINTS', 'POLICY_TYPE', 'TOTAL_CLAIM_AMOUNT', 'VEHICLE_CLASS' ]
df1 = df1.reindex(columns=desired_order)
df2 = df2.reindex(columns=desired_order)
df3 = df3.reindex(columns=desired_order)

display(df1)
display(df2)
display(df3)
```

```python
df_concat = pd.concat([df1, df2, df3])
display(df_concat)
```

```python
print(df_concat.dtypes)
```

```python
df_reduced = df_concat.drop(['EDUCATION', 'NUMBER_OF_OPEN_COMPLAINTS'], axis=1)
display(df_reduced)
```

```python
df_reduced = df_reduced.reset_index(drop=True)
for i in range(len(df_reduced['CUSTOMER_LIFETIME_VALUE'])):
    if isinstance(df_reduced['CUSTOMER_LIFETIME_VALUE'][i], str) and '%' in df_reduced['CUSTOMER_LIFETIME_VALUE'][i]:
        df_reduced['CUSTOMER_LIFETIME_VALUE'][i] = float(df_reduced['CUSTOMER_LIFETIME_VALUE'][i].replace('%', '')) * 100

df_reduced['CUSTOMER_LIFETIME_VALUE'] = pd.to_numeric(df_reduced['CUSTOMER_LIFETIME_VALUE'])

display(df_reduced)
```

```python
df_reduced = df_reduced.drop_duplicates()
display(df_reduced)
```

```python
df_filtered = df_reduced.loc[df_reduced['INCOME'] > 0]
display(df_filtered)
```
